```python
ppeclf.fit(X[:1000,:,:], y[:1000])
```

```python
ppeclf._classification_points
```

```python
test_data_path = "../data/nasa_cmapss/test_FD001.txt"
test = Dataset(test_data_path)
X_test = {i:{} for i in range(ppeclf._classification_points[0], prefix_max_length+1)} # key is number of filled values, subkey is case id
y_test = {i:{} for i in range(ppeclf._classification_points[0], prefix_max_length+1)}
for i, (Xi, yi) in enumerate(test):
    Xi_test = []
    yi_test = []
    # Filter only those without enough filled values
    for j in range(Xi.shape[1]-prefix_max_length+1):
        len_filled = np.sum(~np.isnan(Xi[:, j:j+prefix_max_length])) // Xi.shape[0] # minimum filled values (depending on number of variables)
        print(len_filled)
        if len_filled < ppeclf._classification_points[0]: 
            print(f"Skipping test case due to excessive number of NaNs")
            continue
        else:
            Xi_test.append(Xi[:, j:j+len_filled])
            yi_test.append(yi[j])
            X_test[len_filled][i] = Xi_test
            y_test[len_filled][i] = yi_test
```
